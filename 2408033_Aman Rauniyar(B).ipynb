```python
#Task 3: Outlier Detection
#Identify outliers (1.5  IQR rule):
# Step 1: Calculate IQR for Score and GDP per Capita
Q1 = south_asia_df[['score', 'Log GDP per capita']].quantile(0.25)
Q3 = south_asia_df[['score', 'Log GDP per capita']].quantile(0.75)
IQR = Q3 - Q1

# Define outliers
# Filter the DataFrame to only include the columns used in the IQR calculation
filtered_south_asia_df = south_asia_df[['score', 'Log GDP per capita']]

# Now the comparison will work as both operands have the same columns
outliers = ((filtered_south_asia_df < (Q1 - 1.5 * IQR)) | (filtered_south_asia_df > (Q3 + 1.5 * IQR))).any(axis=1)

# Apply the outlier mask to the original DataFrame
outlier_df = south_asia_df[outliers]

# Print the outlier DataFrame
print(outlier_df)
```

```python
#Scatter plot with outliers:
# Step 3: Scatter plot with outliers highlighted
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], alpha=0.7, label='Non-Outliers')
plt.scatter(outlier_df['Log GDP per capita'], outlier_df['score'], color='red', label='Outliers')
plt.xlabel('Log GDP per capita')  
plt.ylabel('Happiness Score')  
plt.legend()
plt.title('Outlier Detection: Log GDP per capita vs Happiness Score')
plt.show()
```

```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("WHR-2024-5CS037.csv")

# Define the country groups
south_asian_countries = ["Afghanistan", "Bangladesh", "Bhutan", "India", "Maldives", "Nepal", "Pakistan", "Sri Lanka"]
middle_east_countries = ["Bahrain", "Iran", "Iraq", "Israel", "Jordan", "Kuwait", "Lebanon", "Oman",
                         "Palestine", "Qatar", "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"]

# Filter datasets for South Asia and Middle East
south_asia_df = df[df['Country name'].isin(south_asian_countries)]
middle_east_df = df[df['Country name'].isin(middle_east_countries)]
```

```python
#1. Descriptive Statistics:
# Calculate the mean, Standard deviation of the score for both South Asia and Middle East.
# Calculate mean and standard deviation for Score in both regions
sa_mean = south_asia_df['score'].mean()
sa_std = south_asia_df['score'].std()
me_mean = middle_east_df['score'].mean()
me_std = middle_east_df['score'].std()

print("South Asia - Mean Score:", sa_mean, ", Std Dev:", sa_std)
print("Middle East - Mean Score:", me_mean, ", Std Dev:", me_std)

# Which region has higher happiness scores on average?
higher_region = "South Asia" if sa_mean > me_mean else "Middle East"
print(f"The region with higher average happiness scores is: {higher_region}")
```

```python
#2. Top and Bottom Performers
# Identify top 3 and bottom 3 countries based on Score for South Asia
top_3_sa = south_asia_df.nlargest(3, 'score')
bottom_3_sa = south_asia_df.nsmallest(3, 'score')

# Identify top 3 and bottom 3 countries based on Score for Middle East
top_3_me = middle_east_df.nlargest(3, 'score')
bottom_3_me = middle_east_df.nsmallest(3, 'score')

print("Top 3 South Asia Countries:\n", top_3_sa[['Country name', 'score']]) 
print("Bottom 3 South Asia Countries:\n", bottom_3_sa[['Country name', 'score']]) 
print("Top 3 Middle East Countries:\n", top_3_me[['Country name', 'score']])
print("Bottom 3 Middle East Countries:\n", bottom_3_me[['Country name', 'score']]) 

# Bar charts for top and bottom performers
plt.figure(figsize=(10, 6))

# South Asia
```

```python
#3. Metric Comparisons
# Calculate average metrics for GDP, Social Support, and Healthy Life Expectancy
metrics = ['Log GDP per capita', 'Social support', 'Healthy life expectancy']
sa_avg = south_asia_df[metrics].mean()
me_avg = middle_east_df[metrics].mean()

# Plot grouped bar chart
comparison_df = pd.DataFrame({'South Asia': sa_avg, 'Middle East': me_avg})
comparison_df.plot(kind='bar', figsize=(8, 6), color=['skyblue', 'salmon'])
plt.title("Metric Comparison: South Asia vs Middle East")
plt.ylabel("Average Value")
plt.xticks(rotation=0)
plt.show()

# Metric with largest disparity
disparities = abs(sa_avg - me_avg)
largest_disparity = disparities.idxmax()
print(f"Metric with the largest disparity: {largest_disparity}")
```

```python
#Happiness Disparity
# Calculate range and coefficient of variation (CV) for both regions
sa_range = south_asia_df['score'].max() - south_asia_df['score'].min()
me_range = middle_east_df['score'].max() - middle_east_df['score'].min()

sa_cv = (south_asia_df['score'].std() / south_asia_df['score'].mean()) * 100
me_cv = (middle_east_df['score'].std() / middle_east_df['score'].mean()) * 100

print(f"South Asia - Range: {sa_range}, CV: {sa_cv}%")
print(f"Middle East - Range: {me_range}, CV: {me_cv}%")

greater_variability = "South Asia" if sa_cv > me_cv else "Middle East"
print(f"Region with greater variability in happiness: {greater_variability}")
```

```python
#Correlation Analysis
# Correlation of Score with Freedom to Make Life Choices and Generosity
for region, df in zip(['South Asia', 'Middle East'], [south_asia_df, middle_east_df]):
    print(f"\n{region} Correlations:")
    for metric in ['Freedom to make life choices', 'Generosity']:
        correlation = df['score'].corr(df[metric])
        print(f"  {metric}: {correlation}")

        # Scatter plot with trendline
        sns.lmplot(data=df, x=metric, y='score', aspect=2, height=5)
        plt.title(f"{region}: Score vs {metric}")
        plt.show()
```

```python
#Outlier Detection
# Define outliers using 1.5*IQR rule
def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = data[(data[column] < Q1 - 1.5 * IQR) | (data[column] > Q3 + 1.5 * IQR)]
    return outliers
sa_outliers = detect_outliers(south_asia_df, 'score')
me_outliers = detect_outliers(middle_east_df, 'score')

# Print outliers, ensuring column names match
print("South Asia Outliers:\n", sa_outliers[['Country name', 'score']])
print("Middle East Outliers:\n", me_outliers[['Country name', 'score']])

# Plot outliers
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], label='South Asia', alpha=0.6)
plt.scatter(me_outliers['Log GDP per capita'], me_outliers['score'], color='red', label='Middle East Outliers')
plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.legend()
plt.title("Outlier Detection: Log GDP per capita vs score")
plt.show()
```
